```python
from naive_bayes import *
from utils import *

# https://archive.ics.uci.edu/ml/datasets/Congressional+Voting+Records 
showRows('house-votes-train.tsv', 10)
```

```python
c = Classifier('house-votes-train.tsv')
print c.counts
```

```python
x_labels = c.counts.get('republican').keys()
republican_yes = [c.counts['republican'][vote]['y'] for vote in x_labels]
republican_no = [c.counts['republican'][vote]['n'] for vote in x_labels]
democrat_yes = [c.counts['democrat'][vote]['y'] for vote in x_labels]
democrat_no = [c.counts['democrat'][vote]['n'] for vote in x_labels]

print republican_yes
print republican_no
print democrat_yes
print democrat_no
```

```python
import plotly.plotly as py
from plotly.graph_objs import *

trace1 = Bar(
    x=x_labels,
    y=republican_yes,
    name='Republican Yes'
)
trace2 = Bar(
    x=x_labels,
    y=republican_no,
    name='Republican No'
)
trace3 = Bar(
    x=x_labels,
    y=democrat_yes,
    name='Democrat Yes'
)
trace4 = Bar(
    x=x_labels,
    y=democrat_no,
    name='Democrat No'
)
data = Data([trace1, trace2, trace3, trace4])
layout = Layout(
    barmode='stack'
)
fig = Figure(data=data, layout=layout)
```

```python
py.iplot(fig, filename='votes-numbers')
```

```python
# Looking at the probability of a "yes" vote for each item belonging to a Democrat
democrat_yes_prob = [c.conditional['democrat'][vote]['y'] for vote in x_labels]

trace1 = Bar(
    x=x_labels,
    y=democrat_yes_prob,
    name='Democrat Yes : Probability'
)

data = Data([trace1])
py.iplot(data, filename='democrat-yes-prob')
```
