```python
filename = "clustering/new_cereals.csv"
showRows(filename, 10)
```

```python
clusterer = hClusterer(filename)
cluster = clusterer.cluster()

printDendrogram(cluster)
```

```python
import plotly.plotly as py
from plotly.graph_objs import *

data = Data([
    Heatmap(
        z=clusterer.distances,
        x=clusterer.labels,
        y=clusterer.labels,
        colorscale='Portland',
        reversescale=True 
    )
])

width = 600   # plot width 
height = 600  # plot height

layout = Layout(
    title='Cereal Clustering',
    autosize=False,  # custom size
    height=height,   # plot's height in pixels 
    width=width      # plot's width in pixels
)
fig = Figure(data=data, layout=layout)

py.iplot(fig, filename='cereal-heatmap')
```
